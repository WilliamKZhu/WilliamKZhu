```python
#importing packages
import requests
import json
import pandas as pd
#unique weather api key
key = "d0a84a46857f98282dd839475b0a4e5b"
#10 cities we pass by when we drive from New York to Buffalo
cities=['New+york',"Newark","East+Stroudsburg","Tobyhanna","Scranton",
        'Binghamton',"Corning","Dansville",'Geneseo','Buffalo']
units= "imperial"
min_temp=[]
max_temp=[]
humidity=[]
visibility=[]
weather=[]
route=[]
```

```python
for city in cities:
    url = "https://api.openweathermap.org/data/2.5/weather?q=" + city +"&units=" + units +"&APPID=" + key
    r = requests.get(url) #getting request from api
    data= r.json() 
    if (r.status_code==200):
        print(json.dumps(data , sort_keys=True , indent= 10)) #overview of api data
        min_temp.append(data["main"]["temp_min"]) #appending data into dataframes
        route.append(data["name"])
        max_temp.append(data["main"]["temp_max"])
        humidity.append(data["main"]["humidity"])
        visibility.append(data["visibility"])
        for i in range(0, len(data["weather"])):
            weather.append(data["weather"][i]['description'])
```

```python
#creating pandas dataframe
df= pd.DataFrame()
#appending pieces of information into the dataframe
df["Min Temperature"]= min_temp
df["Max Temperature"] = max_temp
df["Humidity"] =humidity
df["Visibility"] = visibility
df["Weather Description"] = weather
df["City"] = route
#setting index as city, permanetly 
df.set_index("City",inplace=True)
#printing dataframe with 10 cities and 5 
df
#statistical data on the data frame
df.describe()
```
