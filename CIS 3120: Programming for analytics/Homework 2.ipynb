```python
# importing packages
import pandas as pd
import requests
from bs4 import BeautifulSoup
url= "https://en.wikipedia.org/wiki/Statistics_of_the_COVID-19_pandemic_in_the_United_States"
page = requests.get(url) 
soup= BeautifulSoup(page.content,"html.parser")
```

```python
#scraping the tag 'table' with a class of wikitable.
table= soup.find("table", class_= "wikitable") 

#finding all of the rows inside the variable table
rows= table.findAll('tr')

#creating a matrix for each column
State=[]
Cases=[]
Deaths=[]
Recov=[]
Hosp=[]

# creating a loop for each row of table data
for row in rows:
    cells = row.findAll('td')
    th = row.findAll('th')
###one row of data has a length of 5. defining the loop and then appending them into respective columns
    if len(cells) ==5:
###information on state is located in the table heading and NOT table data
        State.append(th[1].find(text=True).rstrip("\n")) 
        Cases.append(cells[0].find(text=True).rstrip("\n"))
        Deaths.append(cells[1].find(text=True).rstrip("\n"))
        Recov.append(cells[2].find(text=True).rstrip("\n"))
        Hosp.append(cells[3].find(text=True).rstrip("\n"))
        print(cells)
        print(th)
```

```python
#filling in the blanks of the matrix with the data we scrapped
df = pd.DataFrame()
df['States']= State
df['Cases']= Cases
df['Deaths']= Deaths
df['Recoveries']= Recov
df['Hospitalizations']= Hosp
df
```

```python
#replacing string values as integer values to use .describe function
df["Cases"]=df["Cases"].str.replace(",","")
df["Deaths"]=df["Deaths"].str.replace(",","")
df["Cases"]=df["Cases"].astype(int)
df["Deaths"]=df["Deaths"].astype(int)
df.dtypes
```

```python
#using describe function to show quality of data.
df[['Cases','Deaths']].describe()
```
